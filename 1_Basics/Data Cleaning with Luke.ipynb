```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_jobs.csv')
df.head()
```

```python
df.info()
```

```python
df.iloc[90:100]
```

```python
df[(df['job_title_short'] == 'Data Analyst') & (df['salary_year_avg'] > 100000)].head()
```

```python
df.info()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
df = df.sort_values('job_posted_date')
```

```python
df = df.drop(columns='salary_hour_avg')
```

```python
df = df.dropna(subset=['salary_year_avg'])
```

```python
df.describe()
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=True)
job_salary
```

```python
job_salary.plot(kind='barh')
plt.title('Median Salary by Job Title')
plt.xlabel('Salary ($USD)')
plt.ylabel('Job Title')
plt.show()
```

```python
df[df['job_country'] == 'United States'].head(5)
```

```python
job_counts = df['job_title_short'].value_counts()
job_counts
```

```python
type(job_counts)
```

```python
job_counts = job_counts.sort_values(ascending=True)
job_counts.plot(kind='barh')
plt.title('Posting by Job Title')
plt.xlabel('Counts')
plt.ylabel('Job Title')
plt.show()
```

```python
plt.barh(job_counts.index, job_counts)
plt.show()
```

```python
job_months = df['job_posted_month'].value_counts().sort_index()
job_months
```

```python
type(job_months)
```

```python
plt.plot(job_months.index, job_months)
plt.show()
```

```python
plt.figure(figsize=(7, 3))
job_months.plot(kind='line')
plt.title('Job Month Series')
plt.xlabel('Job Month Index')
plt.ylabel('Job Month Values')
plt.show()
```
